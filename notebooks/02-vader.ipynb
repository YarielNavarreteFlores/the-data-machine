```python
# =========================================
# ANÁLISIS DE SENTIMIENTO CON VADER
# =========================================

import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# =========================================
# CARGAR DATASET
# =========================================

df = pd.read_csv(
    #r"C:\Users\Atala Sojo\OneDrive\Escritorio\ProyectoVD\games_march2025_cleaned.csv"
    r"C:\Users\Melanie\Desktop\daad\the-data-machine\data\raw\games_march2025_cleaned.csv"
)

# =========================================
# FILTRAR REVIEWS CON TEXTO
# =========================================

df_reviews = df[
    df['reviews'].notnull() &
    (df['reviews'].astype(str).str.strip() != "")
].copy()

print("\nCantidad de juegos con reviews:")
print(len(df_reviews))

# =========================================
# INICIALIZAR VADER
# =========================================

analyzer = SentimentIntensityAnalyzer()

# =========================================
# FUNCIÓN DE SENTIMIENTO
# =========================================

def analizar_sentimiento(texto):

    score = analyzer.polarity_scores(str(texto))

    return score['compound']

# =========================================
# APLICAR VADER
# =========================================

df_reviews['sentimiento_vader'] = df_reviews['reviews'].apply(
    analizar_sentimiento
)

# =========================================
# CLASIFICAR SENTIMIENTO
# =========================================

def clasificar_sentimiento(valor):

    if valor >= 0.05:
        return "Positivo"

    elif valor <= -0.05:
        return "Negativo"

    else:
        return "Neutral"

df_reviews['clasificacion_vader'] = df_reviews[
    'sentimiento_vader'
].apply(clasificar_sentimiento)

# =========================================
# MOSTRAR RESULTADOS
# =========================================

print("\n========== RESULTADOS VADER ==========\n")

print(
    df_reviews[
        [
            'name',
            'sentimiento_vader',
            'clasificacion_vader'
        ]
    ].head(10)
)

# =========================================
# VALIDACIÓN CONTRA pct_pos_total
# =========================================

print("\n========== VALIDACIÓN CON pct_pos_total ==========\n")

comparacion = df_reviews[
    [
        'name',
        'pct_pos_total',
        'positive',
        'negative',
        'sentimiento_vader',
        'clasificacion_vader'
    ]
]

print(comparacion.head(20))

# =========================================
# CORRELACIÓN
# =========================================

correlacion = df_reviews[
    'sentimiento_vader'
].corr(
    df_reviews['pct_pos_total']
)

print("\n========== CORRELACIÓN ==========\n")

print(
    "Correlación entre VADER y pct_pos_total:",
    correlacion
)

# =========================================
# PROMEDIOS
# =========================================

print("\n========== PROMEDIOS ==========\n")

print(
    "Promedio sentimiento VADER:",
    df_reviews['sentimiento_vader'].mean()
)

print(
    "Promedio pct_pos_total:",
    df_reviews['pct_pos_total'].mean()
)

# =========================================
# CONTEO DE SENTIMIENTOS
# =========================================

print("\n========== CONTEO DE SENTIMIENTOS ==========\n")

print(
    df_reviews['clasificacion_vader'].value_counts()
)

# =========================================
# TOP POSITIVOS SEGÚN VADER
# =========================================

print("\n========== TOP POSITIVOS SEGÚN VADER ==========\n")

top_positivos = df_reviews.sort_values(
    by='sentimiento_vader',
    ascending=False
)

print(
    top_positivos[
        [
            'name',
            'sentimiento_vader',
            'pct_pos_total'
        ]
    ].head(10)
)

# =========================================
# TOP NEGATIVOS SEGÚN VADER
# =========================================

print("\n========== TOP NEGATIVOS SEGÚN VADER ==========\n")

top_negativos = df_reviews.sort_values(
    by='sentimiento_vader',
    ascending=True
)

print(
    top_negativos[
        [
            'name',
            'sentimiento_vader',
            'pct_pos_total'
        ]
    ].head(10)
)

# =========================================
# GUARDAR RESULTADOS
# =========================================

df_reviews.to_csv(
    r"C:\Users\Melanie\Desktop\daad\the-data-machine\data\raw\games_march2025_cleaned.csv",
    #r"C:\Users\Atala Sojo\OneDrive\Escritorio\ProyectoVD\analisis_sentimiento_vader.csv",
    index=False
)

print("\nArchivo guardado correctamente.")

# =========================================
# FINAL
# =========================================

print("\n========== ANÁLISIS COMPLETADO ==========\n")
```
