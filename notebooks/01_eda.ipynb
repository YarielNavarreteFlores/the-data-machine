```python
# =========================================
# EDA DEL DATASET games_march2025_cleaned
# =========================================

import pandas as pd

# =========================================
# CARGAR DATASET
# =========================================

df = pd.read_csv(
    #r"C:\Users\Atala Sojo\OneDrive\Escritorio\ProyectoVD\reviews_con_texto.csv"
    r"C:\Users\Melanie\Desktop\daad\the-data-machine\data\raw\games_march2025_cleaned.csv"
)

# =========================================
# PRIMERAS FILAS
# =========================================

print("\n========== PRIMERAS FILAS ==========\n")
print(df.head())

# =========================================
# COLUMNAS
# =========================================

print("\n========== COLUMNAS ==========\n")
print(df.columns)

# =========================================
# INFORMACIÓN GENERAL
# =========================================

print("\n========== INFORMACIÓN GENERAL ==========\n")
print(df.info())

# =========================================
# ESTADÍSTICAS GENERALES
# =========================================

print("\n========== ESTADÍSTICAS GENERALES ==========\n")
print(df.describe(include='all'))

# =========================================
# VALORES NULOS
# =========================================

print("\n========== VALORES NULOS ==========\n")
print(df.isnull().sum())

# =========================================
# TOP 10 JUEGOS CON MÁS REVIEWS
# usando num_reviews_total
# =========================================

print("\n========== TOP 10 JUEGOS CON MÁS REVIEWS ==========\n")

top_reviews = df.sort_values(
    by='num_reviews_total',
    ascending=False
)

print(top_reviews[['name', 'num_reviews_total']].head(10))

# =========================================
# TOP 10 JUEGOS MEJOR CALIFICADOS
# usando pct_pos_total
# =========================================

print("\n========== TOP 10 JUEGOS MEJOR CALIFICADOS ==========\n")

top_score = df.sort_values(
    by='pct_pos_total',
    ascending=False
)

print(top_score[['name', 'pct_pos_total']].head(10))

# =========================================
# JUEGOS CON MÁS RECOMENDACIONES
# =========================================

print("\n========== TOP 10 JUEGOS CON MÁS RECOMENDACIONES ==========\n")

top_recommendations = df.sort_values(
    by='recommendations',
    ascending=False
)

print(top_recommendations[['name', 'recommendations']].head(10))

# =========================================
# FILTRAR REVIEWS NO VACÍAS
# usando columna reviews
# =========================================

print("\n========== JUEGOS CON REVIEWS DE TEXTO ==========\n")

reviews_texto = df[
    df['reviews'].notnull() &
    (df['reviews'].astype(str).str.strip() != "")
]

print(reviews_texto[['name', 'reviews']].head(10))

print("\nCantidad de juegos con reviews de texto:")
print(len(reviews_texto))

# =========================================
# GUARDAR CSV FILTRADO
# =========================================

reviews_texto.to_csv(
    #r"C:\Users\Atala Sojo\OneDrive\Escritorio\ProyectoVD\reviews_con_texto.csv",
    r"C:\Users\Melanie\Desktop\daad\the-data-machine\data\raw\games_march2025_cleaned.csv",
    index=False
)

print("\nArchivo filtrado guardado correctamente.")

# =========================================
# JUEGOS GRATIS
# =========================================

print("\n========== JUEGOS GRATIS ==========\n")

gratis = df[df['price'] == 0]

print(gratis[['name', 'price']].head(10))

print("\nCantidad de juegos gratis:")
print(len(gratis))

# =========================================
# JUEGOS MÁS CAROS
# =========================================

print("\n========== TOP 10 JUEGOS MÁS CAROS ==========\n")

caros = df.sort_values(
    by='price',
    ascending=False
)

print(caros[['name', 'price']].head(10))

# =========================================
# JUEGOS CON MÁS HORAS JUGADAS
# =========================================

print("\n========== TOP 10 MÁS JUGADOS ==========\n")

jugados = df.sort_values(
    by='average_playtime_forever',
    ascending=False
)

print(jugados[['name', 'average_playtime_forever']].head(10))

# =========================================
# CANTIDAD DE JUEGOS ÚNICOS
# =========================================

print("\n========== TOTAL DE JUEGOS ==========\n")
print(df['name'].nunique())

# =========================================
# FINAL
# =========================================

print("\n========== EDA COMPLETADO ==========\n")
```
